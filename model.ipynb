#Feature Engineering

```python
data = d2.copy()
data.head()
```

```python
data['Dept_flights_time'].value_counts()
```

```python
# Convert "Dept flights time" datatype from object to integer by using mapping.

data['Dept_flights_time']=data['Dept_flights_time'].map({'Morning_flight':1,'Afternoon_flight':3, 'Evening_flight':3, 'Night_flight': 4}).astype('int')
```

```python
# Convert "Cabin" datatype from object to integer by using mapping.

data['Cabin']=data['Cabin'].map({'E':1,'B':2,'PE':3}).astype('int')
```

```python
data.head(5)
```

#Encoding Categorical Columns

```python
# Encoding the categorical columns

from sklearn.preprocessing import LabelEncoder
encode = LabelEncoder()
categorical_cols = data.select_dtypes(include ='object').columns

for item in categorical_cols:
  data[item] = encode.fit_transform(data[item])
```

```python
data.head(10)
```

LabelEncoder encodes categories into numbers as 0, 1, 2, and so on..


```python
data.info()
```

Now we store all the features into X , by dropping the columns for which we have already created other engineered columns or the columns which we don't require as feature.


```python
data.isnull().sum()
```

```python
data.head()
```

```python
# Dividing our dataset into dependent and independent features

# independent features
x=data.iloc[:, :13]
#x=data.drop('optimal_hours',axis=1)

# dependent feature
y=data.iloc[:, -1]
```

# Feature Scaling
It is done to rescale the features so that they have the properties of standard normal distribution. 

```python
from sklearn.preprocessing import StandardScaler

ss=StandardScaler()
transformed=ss.fit_transform(x)
transformed
```

#PCA

```python
from sklearn.decomposition import PCA
pca=PCA(0.95) # components store 95% of the information
x_pca=pca.fit_transform(transformed)
x_pca.shape
```

```python
# Check the values of eigen vectors
# prodeced by principal components
pca.components_
```

```python
pca.n_components_
```

```python
#Now we know the components are 10 but we can plot 2 or 3d to see their distribution throught visualization 
plt.figure(figsize=(10,10))
plt.scatter(x_pca[:,0],x_pca[:,1],c=y,cmap='plasma')
plt.xlabel('pc1')
plt.ylabel('pc2')
```

```python
# check how much variance is explained by each principal component
print(pca.explained_variance_ratio_)
```

Now Lets do train test splitting to divide the dataset into train and test data for modeling.

#Train Test Splitting

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=101)
```

#Building and Training the model

**Random Forest Regressor**

```python
from sklearn import metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(x_train, y_train)
```

Prediction and Evaluation

```python
rfpredict = rf.predict(x_test)
```

```python
print('Score: ')
rf.score(x_test, y_test)
```

```python
print('MSE:', metrics.mean_squared_error(y_test, rfpredict))
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, rfpredict)))
```

```python
feature_scores = pd.Series(rf.feature_importances_, index=x_train.columns).sort_values(ascending=False)

feature_scores
```

```python
rf.feature_importances_
```

```python
sorted_idx = rf.feature_importances_.argsort()
plt.barh(data.columns[sorted_idx], rf.feature_importances_[sorted_idx])
plt.xlabel("Random Forest Feature Importance")
```

```python
# Saving model to disk
import pickle
pickle.dump(rf, open('model.pkl','wb'))
```

We got the results that we desired . 

So we don't need to fine-tune 'max_depth' and 'n_estimators' parameters for our RandomForest Regressor by using **GridSearch**.
